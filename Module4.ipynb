```python
# Atir Yunus Khan (21BCE2707)

import numpy as np
import statistics
from scipy import stats

data = [2, 4, 4, 4, 5, 5, 7, 9]

# Arithmetic Mean
arithmetic_mean = np.mean(data)

# Median
median = np.median(data)

# Mode
mode = statistics.mode(data)

# Harmonic Mean
harmonic_mean = statistics.harmonic_mean(data)

# Geometric Mean
geometric_mean = stats.gmean(data)

# Quadratic Mean (Root Mean Square)
quadratic_mean = np.sqrt(np.mean(np.square(data)))

# Alpha Mean (for example, alpha = 0.5)
alpha = 0.5
alpha_mean = np.mean([x**alpha for x in data]) ** (1/alpha)

# Quantiles (0.25, 0.5, 0.75 by default)
quantiles = np.quantile(data, [0.25, 0.5, 0.75])

# Quartiles
quartiles = np.percentile(data, [25, 50, 75])

# Deciles (splitting the data into 10 equal parts)
deciles = np.percentile(data, np.arange(0, 101, 10))

# Display the results
print(f"Arithmetic Mean: {arithmetic_mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Harmonic Mean: {harmonic_mean}")
print(f"Geometric Mean: {geometric_mean}")
print(f"Quadratic Mean: {quadratic_mean}")
print(f"Alpha Mean: {alpha_mean}")
print(f"Quantiles (25%, 50%, 75%): {quantiles}")
print(f"Quartiles: {quartiles}")
print(f"Deciles: {deciles}")
```

```python
import numpy as np
import pandas as pd
from scipy import stats

data = [2, 4, 4, 4, 5, 5, 7, 9]

# Range (R)
range_R = np.ptp(data)

# Interquartile Range (IQR)
iqr = stats.iqr(data)

# Mean Difference (average absolute deviation from the mean)
mean_diff = np.mean(np.abs(data - np.mean(data)))

# Standard Deviation
std_dev = np.std(data, ddof=1)

# Variance
variance = np.var(data, ddof=1)

# Skewness Measure
skewness = stats.skew(data)

# Kurtosis (Excess Kurtosis)
kurtosis = stats.kurtosis(data)

# Frequency Distribution (using pandas to create a frequency table)
frequency_distribution = pd.Series(data).value_counts()

# Display the results
print(f"Range (R): {range_R}")
print(f"Interquartile Range (IQR): {iqr}")
print(f"Mean Difference: {mean_diff}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
print("\nFrequency Distribution:")
print(frequency_distribution)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = [2, 4, 4, 4, 5, 5, 7, 9]

# Frequency Distribution
frequency_distribution = pd.Series(data).value_counts().sort_index()

# 1. Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=8, edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 2. Frequency Polygon
plt.figure(figsize=(10, 6))
unique_data = sorted(frequency_distribution.index)
frequencies = frequency_distribution.values

plt.plot(unique_data, frequencies, marker='o', linestyle='-', color='b')
plt.title('Frequency Polygon')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 3. Relative Frequency Distribution (as percentages)
relative_frequencies = frequency_distribution / len(data)
plt.figure(figsize=(10, 6))
plt.bar(frequency_distribution.index, relative_frequencies * 100, color='green', edgecolor='black')
plt.title('Relative Frequency Distribution (as percentages)')
plt.xlabel('Data Values')
plt.ylabel('Relative Frequency (%)')
plt.grid(True)
plt.show()

# 4. Cumulative Frequency Distribution (Ogive)
cumulative_frequencies = frequency_distribution.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(frequency_distribution.index, cumulative_frequencies, marker='o', linestyle='-', color='r')
plt.title('Cumulative Frequency Distribution (Ogive)')
plt.xlabel('Data Values')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()

# 5. Box Plot (to visualize dispersion)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, color='lightblue')
plt.title('Box Plot (Dispersion)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample categorical data
data = ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Apple', 'Banana', 'Orange', 'Apple']

# 1-D Analysis: Frequency distribution of categorical variable
category_counts = pd.Series(data).value_counts()
print("Frequency Distribution:\n", category_counts)

# Pie Chart
plt.figure(figsize=(7, 7))
category_counts.plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Pie Chart of Categorical Variable')
plt.ylabel('')  # Hide ylabel
plt.show()

# Component Part Bar Chart
plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Component Part Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Sample multi-level categorical data
data = pd.DataFrame({
    'Fruits': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Apple', 'Banana', 'Orange', 'Apple'],
    'Region': ['North', 'South', 'West', 'East', 'South', 'West', 'North', 'East', 'North']
})

# 1-D Analysis with multiple forms (Group by multiple categories)
category_group = data.groupby(['Fruits', 'Region']).size().unstack(fill_value=0)
print("Grouped Frequency Distribution:\n", category_group)

# Stacked Bar Chart (multi-level categories)
category_group.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Stacked Bar Chart of Categorical Variable with Multiple Forms')
plt.xlabel('Fruits')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Sample time series data
time_series_data = pd.date_range(start='2023-01-01', periods=10, freq='M')  # Monthly data
values = [100, 110, 120, 130, 125, 135, 140, 145, 150, 155]

# Create DataFrame
time_series_df = pd.DataFrame({'Date': time_series_data, 'Values': values})
time_series_df.set_index('Date', inplace=True)

# 1-D Analysis of Time Series: Summary Statistics
print("Time Series Summary Statistics:\n", time_series_df.describe())

# Line Plot for Time Series
plt.figure(figsize=(10, 6))
plt.plot(time_series_df.index, time_series_df['Values'], marker='o', linestyle='-', color='b')
plt.title('Time Series Analysis (Chronological Variables)')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
